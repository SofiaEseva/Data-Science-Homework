```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import sqrt

x1 = []
x2 = []
y1 = []
y2 = []
a = 2
b = 10
for x_ in range (a, 40):
    x1.append(x_)
    x2.append(-x_)
    y1.append(sqrt(b**2 *(x_**2 - a**2)/ a**2))
    y2.append(-sqrt(b**2 *(x_**2 - a**2)/ a**2))
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y1)
plt.plot(x2, y2)
```

```python
#Нарисуйте трехмерный график двух параллельных плоскостей.
%matplotlib inline
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
x1 = np.arange(-5, 5, 0.5)
y1 = np.arange(-5, 5, 0.5)
x2 = np.arange(-1.5, 5, 0.5)
y2 = np.arange(-2.5, 5, 0.5)
x1, y1 = np.meshgrid(x1, y1)
x2, y2 = np.meshgrid(x2, y2)
z1 = 2*x1 + 5*y1
z2 = 30*x2 - 6*y2
ax.plot_surface(x1, y1, z1)
ax.plot_surface(x2, y2, z2)
show()
```

```python
#Нарисуйте трехмерный график двух любых поверхностей второго порядка.
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)
z = x**2 + y**2
ax.plot_surface(x, y, z)
show()
```
