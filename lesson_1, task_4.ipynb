```python
#Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), 
#взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.cos(x)
y2 = np.cos(3 * x)
line1 = plt.plot(x, y1,label='Линия 1', color = '#20B2AA', linestyle = '--')
line2 = plt.plot(x, y2,label='Линия 2', color = '#DA70D6', linestyle = ':')
plt.legend()



```
